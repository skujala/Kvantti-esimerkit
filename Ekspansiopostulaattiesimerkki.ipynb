### Tehtävänanto 

Äärettömässä potentiaalikaivossa (leveys $a$) sijaitsevalla hiukkasella on ajanhetkellä $t=0$ aaltofunktio muotoa 
$$\Psi(x,0) =
   \begin{cases}
      Ax & 0\leq x \leq L/2\\
      A(L-x) & L/2 \leq x \leq L,
    \end{cases}
$$
missä $A={2\sqrt 3 \over \sqrt{L^3}}$.

### Ratkaisu osa 1

Tehtävänannossa pyydetään tutkimaan äärettömässä $a$-levyisessä potentiaalikaivossa olevaa hiukkasta, joka on preparoitu ajanhetkellä $t=0$ tehtävänannossa annettuun tilaan $\Psi(x,0)$. Tehtävän tavoitteena on tutkia aaltofunktion aikakäyttäytymistä, eli _temporaalista evoluutiota_. Kuvasta nähdään selvästi, että aaltofunktio ei ole muodoltaan samanlainen kuin äärettömän kvanttikaivon ominaisfunktiot $$\Psi_n(x,t) = \sqrt{2\over L} \sin {n\pi x\over L}\,\exp{i\omega_n t},$$ kun $\omega_n = E_n/\hbar$ ja $E_n = {n^2 \pi^2 \hbar^2 \over 2mL^2}$.

Tällöin annettu aaltofunktio voidaan ekspansiopostulaatin avulla kirjoittaa painotettuna summana $$\Psi(x,t) = \sum_i c_i \Psi_n(x,t) \text{ missä } c_i = \int\limits_{-\infty}^{\infty} \Psi_n^\ast \Psi dx.$$ Nämä integraalit ovat tyypillisesti ikäviä laskea käsin, usein jopa mahdottomia. Tällöin paras ratkaisu on antaa tietokoneen ratkaista tämäkin probleema (laskuharjoitus 4).

HUOM! Vektorimaailman analogia ekspansiopostulaatille on vektorin $\vec{A}$ esittämiselle yksikkö- tai kantavektorien $\hat{e}_1$, $\hat{e}_2$ ja $\hat{e}_3$ avulla: $\vec{A} = \sum_n c_n \hat{e}_n$, jossa $c_n = \hat{e}_n \cdot \vec{A}$. Ekspansiopostulaatissa kantavektorit muuttuvat kantafunktioiksi ja piste- tai sisätulo (inner product) integroinniksi. Itseasiassa integraalia $\int\limits_{-\infty}^{\infty} \Psi_n^\ast \Psi dx$ kutsutaan sitäkin sisätuloksi (vrt. Fourier-kurssi ja Fourier-sarjakehitelmä).

#### Aaltofunktion normalisointi

Jotta aaltofunktio olisi todennäköisyysamplitudi, sen itseisarvon neliön (todennäköisyystiheys) integraali kvanttikaivon yli on oltava 1:
$\int\limits_0^L \left|\Psi(x,0)\right|^2 dx = 1.$ Tästä saadaan laskettua (kokeile itse!) $A$:n arvoksi $A={2\sqrt{3}\over\sqrt{L^3}}$.

$$\begin{align*}
\int\limits_0^L \left|\Psi(x,0)\right|^2 dx &= A^2 \left(\int\limits_0^{L/2} x^2 dx + \int\limits_{L/2}^L (L-x)^2 dx\right) = \\
&= A^2\left(2{1\over 3}\Big[{L\over 2}\Big]^3\right) = A^2\left({2\over 3}{L^3 \over 8}\right) = A^2\left({1\over 3}{L^3 \over 4}\right) = 1 \\
&\implies A^2 = {12\over L^3} \implies A = {2\sqrt{3}\over L^{3/2}}
\end{align*}$$

#### Aaltofunktion aikaevoluutio

Aaltofunktion aikaevoluutio saadaan ekspansiopostulaatin avulla. Tätä varten tarvitaan tieto tehtävän ominaisfunktioista. Koska kyseessä on ääretön kvanttikaivo, sen normeeratut ominaisfunktiot ovat 

Tällöin ekspansiopostulaatin mukaan hiukkasta kuvaava aaltofunktio voidaan esittää:

$$\begin{align*}
c_n &= \sqrt{\frac{2}{L}}\frac{2\sqrt{3}}{\sqrt{L^3}}\left( \int_0^{L/2} x\sin\left(\frac{n\pi x}{L}\right)+ \int_{L/2}^{L} (L-x)\sin\left(\frac{n\pi x}{L}\right) dx \right) \\
&=\frac{2\sqrt{6}}{L^2} \Bigg(  \left. \left[ \left(\frac{L}{n\pi}\right)^2 \sin \left(\frac{n\pi x}{L}\right)-\frac{xL}{n\pi}\cos \left(\frac{n\pi x}{L}\right) \right] \right|_0^{L/2}+
\left. L\left[ -\frac{L}{n\pi}\cos \left(\frac{n\pi x}{L}\right) \right] \right|_{L/2}^{L}\\ 
&\phantom{=} -
\left. \left[ \left(\frac{L}{n\pi}\right)^2 \sin \left(\frac{n\pi x}{L}\right)-\frac{xL}{n\pi}\cos \left(\frac{n\pi x}{L}\right) \right] \right|_{L/2}^{L} \Bigg) \\ 
&=\frac{2\sqrt{6}}{L^2} \Bigg[\left(\frac{L}{n\pi}\right)^2 \sin\left(\frac{n\pi}{2}\right) - {\frac{L^2}{2n\pi}\cos\left(\frac{n\pi}{2}\right)}-{\frac{L^2}{n\pi}\cos n\pi}+{\frac{L^2}{n\pi}\cos\left(\frac{n\pi}{2}\right)} \\ 
&\phantom{=}+
\left(\frac{L}{n\pi}\right)^2 \sin\left(\frac{n\pi}{2}\right)+{\frac{L^2}{n\pi}\cos n\pi}-{\frac{L^2}{2n\pi}\cos\left(\frac{n\pi}{2}\right)} \Bigg] \\ 
&= \frac{2\sqrt{6}}{{L^2}}\,2\frac{{L^2}}{(n\pi)^2}\sin\left(\frac{n\pi}{2}\right)= \frac{4\sqrt{6}}{(n\pi)^2}\sin \left(\frac{n\pi}{2}\right) =  \left\{  \begin{array}{l l}
   0& \quad n\;\mathrm{parillinen}\\
  (-1)^{(n-1)/2}\frac{4\sqrt 6}{(n\pi)^2}& \quad  n\;\mathrm{pariton} \,.
\end{array} \right.
\end{align*}$$

Näin saadaan siis 
$$\begin{align*}
c_n &= \left\{  \begin{array}{l l}
   0& \quad n\;\mathrm{parillinen}\\
  (-1)^{(n-1)/2}\frac{4\sqrt 6}{(n\pi)^2}& \quad  n\;\mathrm{pariton} \,.
\end{array} \right.\\
\Psi(x,t) &= \sum_{n=1,3,5,\dots} c_n \sin \left(\frac{n\pi x}{L}\right) \exp \left( -i E_n t /\hbar\right)
\end{align*}$$

```python
# ***********************************************************
# * Nämä funktiot olettavat työtilassa olevan kvanttikaivon *
# * leveyttä kuvaavan muuttujan L                           *
# ***********************************************************

# Alkuperäinen aaltofunktio
def Psi_0(x):
    A = 2 * np.sqrt(3) / pow(L, 3/2)
    
    return np.piecewise(x, 
       [x < 0, (x >=0) & (x < L/2), (x>= L/2) & (x <= L), x > L], 
       [lambda x: 0, lambda x: A*x, lambda x: A*(L-x), lambda x: 0]
    )
```

```python
# Kantafunktio
def Psi_n(n,x):
    return np.sqrt(2/L) * np.sin(n*np.pi / L * x)

# Sarjakehitelmän kertoimet -- laskettiin yllä.
# Monimutkaisemmassa tapauksessa voidaan laskea tässä 
# numeerisella integroinnilla
def cn(n):
    return 4*np.sqrt(6) / (n*np.pi)**2 * np.sin(n*np.pi/2)
```

```python
# Generaattorifunktio (python generator)
# 
# Näppärällinen tapa tutkia jokaisen sarjan termien 
# vaikutusta loppusummaan.
# 
# Kuten aina, tämänkin voi tehdä monella eri tavalla
def expanded_wavefunction(nmax, x):
    result = np.zeros_like(x)
    
    for n in np.arange(1,nmax+1):
        result = result + cn(n)*Psi_n(n, x)
        
        yield result
        
```

### Aikaevoluution tutkiminen

Tehtävänä oli kuitenkin tutkia aineaallon aikaevoluutiota. Siihen päästään muistamalla että jokainen kantafunktio sisältää myös aikakäyttäytymisen. Muodostetaan uusi funktio, jossa tämä on mukana ja plottaillaan todennäköisyystiheyden käyttäytymistä ajan suhteen.

```python
%%capture
def create_expanded_wavefunction_time(nmax, x, t):
    result = np.zeros_like(x)
    
    # Tässä on valittu yksiköt siten, että m = 1 ja \hbar = 1
    
    for n in np.arange(1,nmax+1):
        E_n = n**2*np.pi**2/(200*L**2)
        result = result + cn(n) * Psi_n(n, x) * np.exp(-1j*E_n*t)
        
        
    return np.abs(result)**2
```

```python
L = 1 # Kaivon leveys
x = np.linspace(0, L, 1000)

n_max = 50

# Pohjatyötä animaatiota varten

fig2, ax2 = plt.subplots()
ax2.set_xlim((0, L))
ax2.set_ylim((-.05, 1.05*np.max(Psi_0(x)**2)))

ax2.plot(x, Psi_0(x)**2)

line2, = ax2.plot([], [], lw=2)
```

```python
# initialization function: plot the background of each frame
def init2():
    line2.set_data([], [])
    return (line2,)


# animation function. This is called sequentially
def animate2(i):
    y = create_expanded_wavefunction_time(n_max, x, i)
    line2.set_data(x, y)
    return (line2,)


anim2 = animation.FuncAnimation(fig2, animate2, init_func=init2,
                                frames=100, interval=250, blit=True)
```

### Ratkaisu osa 3

Viimeisessä osassa oli tutkittiin energian odotusarvoa $\left<E\right>$, joka kertoo mikä on todennäköisin arvo hiukkasen energialle, kun se mitataan. Odotusarvo on toiselta nimeltään keskiarvo ja se saadaan ekspansiopostulaatin avulla koska äärettömän kvanttikaivon energian ominaisarvot tunnetaan $\displaystyle\Bigl(E_n = {n^2\pi^2\hbar^2 \over 2mL}\Bigr)$, jolloin 
$\displaystyle\left<E\right> = \sum_n c_n E_n$.

Perustellaan se vielä:
$$
\begin{align*}
\left<E\right> &= \int \Psi^{*}\, \hat{H} \,\Psi dx = \\
 &=\int \Bigl( \sum_m c_m \psi_m e^{-i E_m t/ \hbar } \Bigr)^\ast \underbrace{\hat{H} \Bigl(\sum_n  c_n \psi_n e^{-i E_n t/ \hbar } \Bigr)}_{\sum  c_n E_n \psi_n e^{-i E_n t/ \hbar }}dx \\
&=\sum_m \sum_n c_m^{*}  c_n E_n e^{-i (E_n-E_m)t / \hbar }\underbrace{\int \psi_m^{*} \psi_n dx}_{\delta_{mn}}
 \\&= \sum_n |c_n|^2 E_n \,.
\end{align*}
$$

Tästä saadaan laskettua energian odotusarvo (ikävännäköinen sarja...)
$$
\left<E\right> = \sum_n |c_n|^2 E_n  = \frac{96}{\pi^4}\frac{\pi^2\hbar^2}{2mL^2} \underbrace{\left( \frac{1}{1}+\frac{1}{3^2}+\frac{1}{5^2}+\frac{1}{7^2}+\dots\right)}_{\pi^2 / 8}= \frac{6\hbar^2}{mL^2}
$$

Kun valitaan yksiköt s.e., $\hbar = 1$, $L=1$ ja $m=1$, energian odotusarvo saadaan esitettyä tietokoneen kannalta mukavammassa muodossa
$$\left<E\right> = 6.$$

Katsotaan mitä mieltä asiasta ollaan laskennallisesti:

```python
sum([np.abs(cn(n))**2 * n**2 * np.pi**2 / (2 * L**2) for n in np.arange(1,n_max+1)])
```
